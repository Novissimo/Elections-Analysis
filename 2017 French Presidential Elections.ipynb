In April, and again in May, the French went to the polls to choose a new president. The 2017 elections marked a crucial historical moment for France and Europe. Like the votes in America and Britain in 2016, the French presidential race was a contest between far right nativists and liberal centrists.

If the nativists won, then France may have withdrawn from the European Union, triggering its collapse. If the liberals won, France would work towards maintaining and strengthening the EU and help stem the rising tide of worldwide reactionary nationalism.

The nativist in this case, Marine Le Pen, leader of the National Front and daughter of  long time far right politician Jean-Marie Le Pen, is an “economic nationalist” and Euroskeptic with anti-immigrant views who believes multiculturalism has failed.

The liberal, Emmanuel Macron, was the former Minister of Economy and Finance in the cabinet of President Hollande, who seeks greater European integration and supports an open door to refugees from Africa and the Middle-east.

With so much at stake, the election captured the attention of the world, though on closer inspection the results were not too surprising.

French elections have two rounds. The first round is between any number of candidates. If no candidate wins an outright majority, a second round is held between the two candidates who performed the best in the first round.

```python
#Imports 

#For Data Analysis

import pandas as pd
from pandas import Series,DataFrame
import numpy as np

#For Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

### Averages

```python
#What are the average percentage values b/w candidates?

avg = pd.DataFrame(poll_df.mean())
avg.drop('Number of Observations',axis=0,inplace=True)
avg.drop('Question Iteration', axis=0, inplace=True)
avg.drop('Question Text',axis=0,inplace=True)

#Get standard deviation and check the error

std = pd.DataFrame(poll_df.std())
std.drop('Number of Observations',axis=0,inplace=True)
std.drop('Question Iteration',axis=0,inplace=True)
std.drop('Question Text',axis=0,inplace=True)
#Make a plot
#yerr (y error) set equal to std

french_election_avg = avg.plot(yerr=std,kind='bar',figsize=(10,8),legend=False)
french_election_avg.set_title('Poll Averages for the 2017 French Presidential Election, 1st Round')
french_election_avg.set_ylabel('Poll Averages')
```

In polls that included the top five 2017 candidates, Le Pen did exceptionally well. She led the polls on average. The only other candidate that maintained competitiveness with her was Macron.

```python
##What are the average percentage values b/w candidates during the run offs?

runoff_avg = pd.DataFrame(runoff_df.mean())
runoff_avg.drop('Number of Observations',axis=0,inplace=True)
runoff_avg.drop('Question Iteration',axis=0,inplace=True)
runoff_avg.drop('Question Text',axis=0,inplace=True)

#Get standard deviation and check the error

runoff_std = pd.DataFrame(runoff_df.std())
runoff_std.drop('Number of Observations',axis=0,inplace=True)
runoff_std.drop('Question Iteration',axis=0,inplace=True)
runoff_std.drop('Question Text',axis=0,inplace=True)

#Make a plot 
#yerr (y error) set equal to std

french_runoff_avg = runoff_avg.plot(yerr=runoff_std,kind='bar',figsize=(10,8),legend=False)
french_runoff_avg.set_title('Poll Average for the 2017 French Presidential Election, 2nd Round')
french_runoff_avg.set_ylabel('Poll Average (%)')
```

```python
Interestingly, Macron held a double digit lead over Le Pen in one-on-one polls, where it was just the two of them.
```

```python
#More precise look at average and std

poll_avgs = pd.concat([avg,runoff_avg,std,runoff_std],axis=1)

#Rename colums
poll_avgs.columns = ['1st Round Average','2nd Round Average','1st Round STD','2nd Round STD']

#Show
poll_avgs
```
